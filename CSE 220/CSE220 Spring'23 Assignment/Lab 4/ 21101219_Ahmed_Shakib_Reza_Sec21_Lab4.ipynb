```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
    # To Do
    self.head = Node(a[0],None,None)
    tail = self.head
    for i in range(1,len(a)):
      new_node = Node(a[i],None,tail)
      tail.next = new_node
      tail = tail.next
    tail.next = self.head
    self.head.prev = tail



  
  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    c = 1
    tail = self.head
    while tail.next != self.head:
      c+=1
      tail = tail.next
    return c
  
  # prints the elements in the list
  def forwardprint(self):
    # To Do

    tail = self.head
    while tail.next != self.head:
      print(tail.element,end=' ')
      tail = tail.next
    print(tail.element,end=' ')
    print()

  # prints the elements in the list backward
  def backwardprint(self):
    # To Do

    tail = self.head.prev
    while tail.prev != self.head.prev:
      print(tail.element,end=' ')
      tail = tail.prev
    print(tail.element,end=' ')
    print()

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    c = 0
    tail = self.head
    while tail.next != self.head:
      if c != idx:
        c+=1
        tail = tail.next
      else:
        return tail
    print("index error")
    
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    c = 0
    tail = self.head
    while tail.element != elem:
        c+=1
        tail = tail.next
 

    if c < self.countNode():
      return c
    else:
      return -1

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    # To Do
    c = 0
    tail = self.head
    if c < self.countNode():
      while tail.next != self.head:
        if idx == 0:
          n = Node(elem,self.head,self.head.prev)
          self.head.prev.next = n
          self.head.prev = n
          self.head = n
          return
        else:
          if c!=idx-1:
            c+=1
            tail = tail.next
          else:
            n = Node(elem,None,tail)
            n.next = tail.next
            tail.next.prev = n
            tail.next = n
            return
      n = Node(elem,None,tail)
      n.next = tail.next
      tail.next.prev = n
      tail.next = n
            
    else:
      return None



    
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    c = 0
    tail = self.head
    if c < self.countNode():
      while tail.next != self.head:
        if idx == 0:
          tail.prev.next = tail.next
          tail.next.prev = tail.prev
          self.head = tail.next
          return str(tail.element)
          break
        else:
          if c!=idx:
            c+=1
            tail = tail.next
          else:
            tail.prev.next = tail.next
            tail.next.prev = tail.prev
            return str(tail.element)
            break
      tail.prev.next = tail.next
      tail.next.prev = tail.prev
      return str(tail.element) 
    else:
      return None  
```
