```python
#TASK 1  

def secdule(arr):
  arr2 = []
  arr2.append(arr[0])
  for x in range(1,len(arr)):
    for elem in arr2:
      if arr[x][0] >= elem[1]:
        flag = True
      else:
        flag = False
        break
    if flag == True:
      arr2.append(arr[x])
  output=open("output1.txt","w")
  output.writelines(str(len(arr2)))
  output.writelines("\n")
  for i in range(len(arr2)):
    output.writelines(str(arr2[i][0]))
    output.writelines(" ")
    output.writelines(str(arr2[i][1]))
    output.writelines("\n")
  

input =open("input1.txt","r")
arr = []
n = input.readline()
for i in range(int(n)):
  m = input.readline()
  x,y = m.split()
  arr.append([int(y),int(x)])
arr.sort()
for i in range(len(arr)):
  arr[i].reverse()

secdule(arr)
```

```python
#TASK 2

def secdule(arr,person):
  arr2 = []
  arr2.append(arr[0])
  for x in range(1,len(arr)):
    count = 1
    for elem in arr2:
      if arr[x][0] >= elem[1]:
        flag = True

      else:
        flag = False
        count += 1
    if flag == True:
      arr2.append(arr[x])
    elif flag == False and count == person:
      arr2.append(arr[x])
      
  output = open("output2.txt","w")
  output.writelines(str(len(arr2)))

input =open("input2.txt","r")
arr = []
r = input.readline()
n,p = r.split()
for i in range(int(n)):
  m = input.readline()
  x,y = m.split()
  arr.append([int(y),int(x)])
arr.sort()
for i in range(len(arr)):
  arr[i].reverse()

secdule(arr,int(p))
```

```python
#TASK - 3 
input = open("input3.txt","r")
output = open("output3.txt","w")
n = input.readline()
n = int(n)
m = input.readline()
call = input.readline()
arr = []
for elem in m:
  if  elem != " ":
    if elem!='\n':
      elem = int(elem)
      arr.append(elem)
arr.sort()
J = 0 
j = 0 
done_J = [] 
for elem in call:
  if elem == "J":
    hr = arr.pop(0)
    J += hr
    done_J.append(hr)
  if elem == "j":
    hr = done_J.pop()
    j += hr
  
  output.writelines(str(hr))
output.writelines("\n") 
output.writelines("Jack will work for {} hours".format(J))
output.writelines("\n")
output.writelines("Jill will work for {} hours".format(j))
```

```python
#TASK - 4 (final)
import math

input = open("input4.txt","r")
output = open("output4.txt","w")
x = None
y = None
while x != 0 and y != 0:
  count = 0
  n = input.readline()
  x,y = n.split()
  x = int(x)
  y.strip()
  y = int(y)
  for i in range(x,y+1):
    temp = math.sqrt(i)
    if temp != 0:
      temp2 = math.floor(temp)
      if temp == temp2:
        count+=1
  if count != 0:
    output.writelines(str(count))
    output.writelines("\n")
```
