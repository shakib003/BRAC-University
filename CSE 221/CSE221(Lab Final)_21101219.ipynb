```python
from heapq import heapify

def graph():
  input = open("input.txt","r")
  n = input.readline()
  x,y = n.split()
  x = int(x)
  y = y.strip()
  y = int(y)
  g = {}
  v = []
  v2 = []
  for i in range(y):
    line = input.readline()
    a,b,w = line.split("+")
    w = w.strip()
    w = int(w)
    if a not in v:
      g[a] = {b:w}
      v.append(a)
      if a not in v2:
        v2.append(a)
      if b not in v2:
        v2.append(b)
    else:
      g[a].update({b:w})
      if b not in v2:
        v2.append(b)

  for i in v2:
    if i not in g:
      g[i] = {}
  #print(g)
  #print(v)
  #print(v2)
  return g



def dijkstra(s,d,g):
  heap = []
  v = []
  info = {}
  
  for i in g:
    info[i] = [100000,[]]
  info[s][0] = 0
  #info["2"] = [100000,[]]
  
  print(info)
  temp = s
  while temp in g and  temp != d:
    if temp not in v:
      v.append(temp)
    for i in g[temp]:
      if i not in v:
        cost = info[temp][0] + g[temp][i]
        if cost < info[i][0]:
          info[i][0] = cost
        info[i][1] = info[temp][1] + [temp]
        heap.append((info[i][0],i))
    heapify(heap)
    j = heap.pop(0)
    temp = j[1]

  print("distance",info[d][0])
  print("path",info[d][1]+[d])
g = graph()
dijkstra("0","7",g)
    

```
